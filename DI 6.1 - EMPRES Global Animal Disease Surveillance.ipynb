#### Renaming Columns

```python
# Renaming several columns

GADS.rename(columns = {'Id':'ID'}, inplace=True)
GADS.rename(columns = {'latitude':'Latitude'}, inplace=True)
GADS.rename(columns = {'longitude':'Longitude'}, inplace=True)
GADS.rename(columns = {'region':'Region'}, inplace=True)
GADS.rename(columns = {'country':'Country'}, inplace=True)
GADS.rename(columns = {'admin1':'City/Providence'}, inplace=True)
GADS.rename(columns = {'localityName':'Locality'}, inplace=True)
GADS.rename(columns = {'localityQuality':'LocalityQuality'}, inplace=True)
GADS.rename(columns = {'observationDate':'Observe Date'}, inplace=True)
GADS.rename(columns = {'reportingDate':'Report Date'}, inplace=True)
GADS.rename(columns = {'status':'Status'}, inplace=True)
GADS.rename(columns = {'disease':'Disease'}, inplace=True)
GADS.rename(columns = {'serotypes':'Serotypes'}, inplace=True)
GADS.rename(columns = {'speciesDescription':'Sp. Description'}, inplace=True)
GADS.rename(columns = {'sumAtRisk':'Sum Sp.AtRisk'}, inplace=True)
GADS.rename(columns = {'sumCases':'Sum Sp.Cases'}, inplace=True)
GADS.rename(columns = {'sumDeaths':'Sum Sp.Deaths'}, inplace=True)
GADS.rename(columns = {'sumDestroyed':'Sum Sp.Destroyed'}, inplace=True)
GADS.rename(columns = {'sumSlaughtered':'Sum Sp.Slaughtered'}, inplace=True)
GADS.rename(columns = {'humansGenderDesc':'Humans Gender'}, inplace=True)
GADS.rename(columns = {'humansAge':'Humans Age'}, inplace=True)
GADS.rename(columns = {'humansAffected':'Humans Affected'}, inplace=True)
GADS.rename(columns = {'humansDeaths':'Humans Deaths'}, inplace=True)
```

#### Mix Type Data

```python
# Advance Python Syntax/Function to Search Mixed-Type columns

for col in GADS.columns.tolist():
  weird = (GADS[[col]].applymap(type) != GADS[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (GADS[weird]) > 0:
    print (col)
```

```python
# Addressing Mixed Data

GADS['Observe Date'] = GADS['Observe Date'].astype('str')
```

```python
GADS['Serotypes'] = GADS['Serotypes'].astype('str')
```

```python
GADS['Sp. Description'] = GADS['Sp. Description'].astype('str')
```

```python
GADS['Humans Gender'] = GADS['Humans Gender'].astype('str')
```

```python
# Double checking mix type data was corrected

for col in GADS.columns.tolist():
  weird = (GADS[[col]].applymap(type) != GADS[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (GADS[weird]) > 0:
    print (col)
```

```python
# Mixed type data was corrected
```
